```python
# Importing the required libraries
import requests
import pandas as pd
from bs4 import BeautifulSoup as bs
```

```python
# Downloading contents of the web page
url = requests.get('https://www.studymalaysia.com/education/top-stories/list-of-universities-in-malaysia')
data = bs(url.content)
```

```python
print(data.prettify()) 
```

```python
table = data.find('table', class_="full boxed")
df = pd.read_html(str(table))[0]
df
```

```python
df.to_csv('20 Public Universities in Malaysia.csv', index=False)
```
