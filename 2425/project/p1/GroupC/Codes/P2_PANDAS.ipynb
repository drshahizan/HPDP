Step 3 : Load and Display Dataset, Checking on Total Numbers of Rows and Columns

```python
%%time

tracemalloc.start()
start_time = time.perf_counter()
total_rows = 0

filename = list(uploaded.keys())[0]
file_content = uploaded[filename].decode('utf-8')
df_cleaned = pd.read_csv(io.StringIO(file_content), keep_default_na=False)

total_rows = df_cleaned.shape[0]
display(df_cleaned.head(10))
print(f"Total rows: {df_cleaned.shape[0]}")
print(f"Total columns: {df_cleaned.shape[1]}\n\n\n")

current, peak = tracemalloc.get_traced_memory()
end_time = time.perf_counter()
tracemalloc.stop()

execution_time = end_time - start_time
throughput = total_rows / execution_time

print("===================== Performance =====================\n")
print(f"Total rows processed: {total_rows}")
print(f"Code Execution time: {execution_time:.4f} seconds")
print(f"Throughput: {throughput:.2f} rows per second")
print(f"Current memory usage: {current / 10**6:.4f} MB")
print(f"Peak memory usage: {peak / 10**6:.4f} MB")

cpu_usage = psutil.cpu_percent(interval=1)
print(f"CPU usage: {cpu_usage}%")

print("=======================================================")

print("\nTotal time for this cell (Including time to display the performance):")
```

Step 4 : Dividing Products into 4 Categories Based on Price using Pandas

```python
%%time

tracemalloc.start()
start_time = time.perf_counter()

total_rows = df_cleaned.shape[0]

df_price = df_cleaned[df_cleaned["Price"] > 0]

q1 = df_price["Price"].quantile(0.25)
q3 = df_price["Price"].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

df_cleared = df_price[(df_price["Price"] >= lower_bound) & (df_price["Price"] <= upper_bound)]

min_price = df_cleared["Price"].min()
max_price = df_cleared["Price"].max()
print(f"Min Price: {min_price}")
print(f"Max Price: {max_price}")

price_range = (max_price - min_price) / 4
bound1 = min_price + price_range
bound2 = min_price + 2 * price_range
bound3 = min_price + 3 * price_range

group1 = df_price[df_price["Price"] <= bound1].sort_values("Price")
group2 = df_price[(df_price["Price"] > bound1) & (df_price["Price"] <= bound2)].sort_values("Price")
group3 = df_price[(df_price["Price"] > bound2) & (df_price["Price"] <= bound3)].sort_values("Price")
group4 = df_price[df_price["Price"] > bound3].sort_values("Price")

print(f"\nGroup 1 (Budget Friendly Price): {group1.shape[0]} products")
print(f"Group 2 (Affordable Price): {group2.shape[0]} products")
print(f"Group 3 (Mid-Range Price): {group3.shape[0]} products")
print(f"Group 4 (Premium Price): {group4.shape[0]} products")

def print_category_counts(df_group):
    print(f"📊 Category count:")
    category_counts = df_group["Category"].value_counts()
    for category, count in category_counts.items():
        print(f"- {category}: {count} products")

print("\nGroup 1 (Budget Friendly Price):")
display(group1.head(10))
print(f"Total rows: {group1.shape[0]}")
print(f"Total columns: {group1.shape[1]}\n")
print_category_counts(group1)

print("\nGroup 2 (Affordable Price):")
display(group2.head(10))
print(f"Total rows: {group2.shape[0]}")
print(f"Total columns: {group2.shape[1]}\n")
print_category_counts(group2)

print("\nGroup 3 (Mid-Range Price):")
display(group3.head(10))
print(f"Total rows: {group3.shape[0]}")
print(f"Total columns: {group3.shape[1]}\n")
print_category_counts(group3)

print("\nGroup 4 (Premium Price):")
display(group4.head(10))
print(f"Total rows: {group4.shape[0]}")
print(f"Total columns: {group4.shape[1]}\n")
print_category_counts(group4)

current, peak = tracemalloc.get_traced_memory()
end_time = time.perf_counter()
tracemalloc.stop()

execution_time = end_time - start_time
throughput = total_rows / execution_time

print("\n===================== Performance =====================")
print(f"Total rows processed: {total_rows}")
print(f"Code Execution time: {execution_time:.4f} seconds")
print(f"Throughput: {throughput:.2f} rows per second")
print(f"Current memory usage: {current / 10**6:.4f} MB")
print(f"Peak memory usage: {peak / 10**6:.4f} MB")
print(f"CPU usage: {psutil.cpu_percent(interval=1)}%")
print("=======================================================")

print("\nTotal time for this cell (Including time to display the performance):")
```

Step 5 : Filtering Based on 'Total Reviews' to Determine Popularity of Products

```python
%%time

tracemalloc.start()
start_time = time.perf_counter()
total_rows = df_cleaned.shape[0]

Q1 = df_cleaned["Total Reviews"].quantile(0.25)
Q3 = df_cleaned["Total Reviews"].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df_filtered = df_cleaned[(df_cleaned["Total Reviews"] >= lower_bound) &
                         (df_cleaned["Total Reviews"] <= upper_bound)]

min_ratings = df_filtered["Total Reviews"].min()
max_ratings = df_filtered["Total Reviews"].max()
print(f"Minimum Number of Total Reviews: {min_ratings}")
print(f"Maximum Number of Total Reviews: {max_ratings}")

rating_range = round((max_ratings - min_ratings) / 4)
bound1 = min_ratings + rating_range
bound2 = min_ratings + 2 * rating_range
bound3 = min_ratings + 3 * rating_range

group1 = df_filtered[df_filtered["Total Reviews"] <= bound1].sort_values("Total Reviews", ascending=False)
group2 = df_filtered[(df_filtered["Total Reviews"] > bound1) & (df_filtered["Total Reviews"] <= bound2)].sort_values("Total Reviews", ascending=False)
group3 = df_filtered[(df_filtered["Total Reviews"] > bound2) & (df_filtered["Total Reviews"] <= bound3)].sort_values("Total Reviews", ascending=False)
group4 = df_filtered[df_filtered["Total Reviews"] > bound3].sort_values("Total Reviews", ascending=False)

def print_category_counts(df_group):
    print(f"📊 Category count:")
    category_counts = df_group["Category"].value_counts()
    for category, count in category_counts.items():
        print(f"- {category}: {count} products")

for i, (group, name) in enumerate(zip([group1, group2, group3, group4],
                                      ["Least popular", "Below Average Popularity", "Above Average Popularity", "Most popular"]), 1):
    print(f"\nGroup {i} ({name}):")
    display(group.head(10))
    print(f"Total rows: {group.shape[0]}")
    print(f"Total columns: {group.shape[1]}\n")
    print_category_counts(group)

current, peak = tracemalloc.get_traced_memory()
end_time = time.perf_counter()
tracemalloc.stop()

execution_time = end_time - start_time
throughput = total_rows / execution_time

print("\n===================== Performance =====================")
print(f"Total rows processed: {total_rows}")
print(f"Code Execution time: {execution_time:.4f} seconds")
print(f"Throughput: {throughput:.2f} rows per second")
print(f"Current memory usage: {current / 10**6:.4f} MB")
print(f"Peak memory usage: {peak / 10**6:.4f} MB")

cpu_usage = psutil.cpu_percent(interval=1)
print(f"CPU usage: {cpu_usage}%")
print("=======================================================")
```

Step 6 : Ranking Location Based on Market Performance

```python
%%time

tracemalloc.start()
total_rows = df_cleaned.shape[0]

df_location_sales = df_price.groupby("Location").agg(
    Total_Quantity_Sold=('Quantity Sold', 'sum'),
    Average_Price=('Price', 'mean')
).reset_index()

df_location_sales['Market Performance'] = df_location_sales['Total_Quantity_Sold'] * df_location_sales['Average_Price']

df_location_sales = df_location_sales.sort_values("Market Performance", ascending=False)

display(df_location_sales.head(10))

current, peak = tracemalloc.get_traced_memory()
end_time = time.perf_counter()
tracemalloc.stop()
execution_time = end_time - start_time
throughput = total_rows / execution_time
cpu_usage = psutil.cpu_percent(interval=1)

print("\n===================== Performance =====================")
print(f"Total rows processed: {total_rows}")
print(f"Code Execution time: {execution_time:.4f} seconds")
print(f"Throughput: {throughput:.2f} rows per second")
print(f"Current memory usage: {current / 10**6:.4f} MB")
print(f"Peak memory usage: {peak / 10**6:.4f} MB")

cpu_usage = psutil.cpu_percent(interval=1)
print(f"CPU usage: {cpu_usage}%")
print("=======================================================")
```
