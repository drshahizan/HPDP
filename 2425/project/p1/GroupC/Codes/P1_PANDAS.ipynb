Step 1 : Install and Import Libraries

```python
!pip install pandas
import pandas as pd
import re
import time
import tracemalloc
import psutil
import os
```

Step 2 : Upload Execl Files

```python
from google.colab import files
uploaded = files.upload()
```

Step 3 : Load Excel Files into PANDAS DataFrames

```python
fList_pandas = [pd.read_excel(file) for file in uploaded.keys()]
print(f"Total DataFrames in fList_pandas: {len(fList_pandas)}")
```

Step 4 : Load and Display Dataset, Checking on Total Numbers of Rows and Columns

```python
%%time

tracemalloc.start()
start_time = time.perf_counter()
total_rows = 0

fList_pandas_with_category = []

for filename, df in zip(uploaded.keys(), fList_pandas):
    match = re.search(r"\((.*?)\)", filename)
    category = match.group(1) if match else "Unknown"

    df["Category"] = category
    fList_pandas_with_category.append(df)

print(f"Total DataFrames: {len(fList_pandas_with_category)}")

for df in fList_pandas_with_category:
    total_rows += df.shape[0]
    display(df.head(10))
    print(f"Total rows: {df.shape[0]}")
    print(f"Total columns: {df.shape[1]}\n\n\n")

current, peak = tracemalloc.get_traced_memory()
end_time = time.perf_counter()
tracemalloc.stop()

execution_time = end_time - start_time
throughput = total_rows / execution_time

print("===================== Performance =====================\n")
print(f"Total rows processed: {total_rows}")
print(f"Code Execution time: {execution_time:.4f} seconds")
print(f"Throughput: {throughput:.2f} rows per second")
print(f"Current memory usage: {current / 10**6:.4f} MB")
print(f"Peak memory usage: {peak / 10**6:.4f} MB")

cpu_usage = psutil.cpu_percent(interval=1)
print(f"CPU usage: {cpu_usage}%")

print("=======================================================")

print("\nTotal time for this cell(Including time to display the performance):")
```

Step 5 : Combine All DataFrames into One, Checking on Total Numbers of Rows and Columns

```python
%%time

tracemalloc.start()
start_time = time.perf_counter()

df_combined = pd.concat(fList_pandas, axis=0, ignore_index=True)

display(df_combined.head(10))
total_rows = df_combined.shape[0]
print(f"Total rows: {df_combined.shape[0]}")
print(f"Total columns: {df_combined.shape[1]}\n\n\n")

current, peak = tracemalloc.get_traced_memory()
end_time = time.perf_counter()
tracemalloc.stop()

execution_time = end_time - start_time
throughput = total_rows / execution_time

print("===================== Performance =====================\n")
print(f"Total rows processed: {total_rows}")
print(f"Code Execution time: {execution_time:.4f} seconds")
print(f"Throughput: {throughput:.2f} rows per second")
print(f"Current memory usage: {current / 10**6:.4f} MB")
print(f"Peak memory usage: {peak / 10**6:.4f} MB")

cpu_usage = psutil.cpu_percent(interval=1)
print(f"CPU usage: {cpu_usage}%")

print("=======================================================")
print("\nTotal time for this cell (Including time to display the performance):")
```

Step 6 : Standardizing Field with Object/String Data Type into Uppercase

```python
%%time

tracemalloc.start()
start_time = time.perf_counter()
total_rows = df_combined.shape[0]

for col in df_combined.select_dtypes(include='object').columns:
    df_combined[col] = df_combined[col].str.upper()

display(df_combined.head(10))

print(f"Total rows: {df_combined.shape[0]}")
print(f"Total columns: {df_combined.shape[1]}\n")

current, peak = tracemalloc.get_traced_memory()
end_time = time.perf_counter()
tracemalloc.stop()

execution_time = end_time - start_time
throughput = total_rows / execution_time

print("===================== Performance =====================\n")
print(f"Total rows processed: {total_rows}")
print(f"Code Execution time: {execution_time:.4f} seconds")
print(f"Throughput: {throughput:.2f} rows per second")
print(f"Current memory usage: {current / 10**6:.4f} MB")
print(f"Peak memory usage: {peak / 10**6:.4f} MB")

cpu_usage = psutil.cpu_percent(interval=1)
print(f"CPU usage: {cpu_usage}%")

print("=======================================================")
print("\nTotal time for this cell (Including time to display the performance):")
```

Step 7 : Converting 'Total Reviews' into Integer Data Type (rouding to nearest number)

```python
%%time

tracemalloc.start()
start_time = time.perf_counter()
total_rows = df_combined.shape[0]

df_combined['Total Reviews'] = df_combined['Total Reviews'].astype(str)

df_combined['Total Reviews'] = pd.to_numeric(
    df_combined['Total Reviews'].str.replace(r'[^\d.]', '', regex=True),
    errors='coerce'
).round().astype('Int64')

display(df_combined.head(10))

print(f"Total rows: {df_combined.shape[0]}")
print(f"Total columns: {df_combined.shape[1]}\n")

current, peak = tracemalloc.get_traced_memory()
end_time = time.perf_counter()
tracemalloc.stop()

execution_time = end_time - start_time
throughput = total_rows / execution_time

print("===================== Performance =====================\n")
print(f"Total rows processed: {total_rows}")
print(f"Code Execution time: {execution_time:.4f} seconds")
print(f"Throughput: {throughput:.2f} rows per second")
print(f"Current memory usage: {current / 10**6:.4f} MB")
print(f"Peak memory usage: {peak / 10**6:.4f} MB")

cpu_usage = psutil.cpu_percent(interval=1)
print(f"CPU usage: {cpu_usage}%")
print("=======================================================")
print("\nTotal time for this cell (Including time to display the performance):")
```

Step 8 : Converting 'Quatity Sold' into Integer Data Type

```python
%%time

tracemalloc.start()
process = psutil.Process(os.getpid())
cpu_start = process.cpu_times()
start_time = time.perf_counter()

total_rows = df_combined.shape[0]

df_combined['Quantity Sold'] = df_combined['Quantity Sold'].astype(str)

df_combined['Quantity Sold'] = df_combined['Quantity Sold'].str.replace('SOLD', '', regex=False).str.strip()

def clean_quantity(x):
    if 'K' in x:
        try:
            return float(x.replace('K', '').strip()) * 1000
        except:
            return 0
    else:
        cleaned = re.sub(r'[^\d]', '', x)
        try:
            return float(cleaned)
        except:
            return 0

df_combined['Quantity Sold'] = df_combined['Quantity Sold'].apply(clean_quantity)
df_combined['Quantity Sold'] = pd.Series(df_combined['Quantity Sold']).round().astype('Int64')

display(df_combined.head(10))
print(f"Total rows: {df_combined.shape[0]}")
print(f"Total columns: {df_combined.shape[1]}")
print("\n")

current, peak = tracemalloc.get_traced_memory()
end_time = time.perf_counter()
cpu_end = process.cpu_times()
tracemalloc.stop()

execution_time = end_time - start_time
throughput = total_rows / execution_time

user_time = (cpu_end.user - cpu_start.user)
sys_time = (cpu_end.system - cpu_start.system)
total_cpu_time = user_time + sys_time

print("===================== Performance =====================\n")
print(f"Total rows processed: {total_rows}")
print(f"Code Execution time: {execution_time:.4f} seconds")
print(f"Throughput: {throughput:.2f} rows per second")
print(f"Current memory usage: {current / 10**6:.4f} MB")
print(f"Peak memory usage: {peak / 10**6:.4f} MB")

cpu_usage = psutil.cpu_percent(interval=1)
print(f"CPU usage: {cpu_usage}%")
print("=======================================================")

print("\nTotal time for this cell (Including time to display the performance):")
print(f"CPU times: user {user_time * 1000:.0f} ms, sys: {sys_time * 1000:.1f} ms, total: {total_cpu_time * 1000:.0f} ms")
print(f"Wall time: {execution_time:.2f} s")
```

Step 9 : Checking and Handling Missing Values by Replacing 0 for Numeric Fields and N/A for String/Object Fields

```python
%%time

start_time = time.time()
tracemalloc.start()
total_rows = df_combined.shape[0]

print("Initial Dataset:")
display(df_combined.head(10))

numeric_cols = ['Quantity Sold', 'Total Reviews']
for col in numeric_cols:
    df_combined[col] = pd.to_numeric(df_combined[col], errors='coerce')

df_combined['Quantity Sold'] = df_combined['Quantity Sold'].replace(0, pd.NA)


print("\nBefore Handling Missing Values:")
missing_values = df_combined.isnull().sum()
print(missing_values)

df_combined = df_combined.apply(
    lambda col: col.fillna('N/A') if col.dtype == 'object' else col.fillna(0),
    axis=0
)

print("\nAfter Handling Missing Values:")
missing_values_after = df_combined.isnull().sum()
print(missing_values_after)

print("\nFinalised Dataset:")
display(df_combined.head(10))
print(f"Total rows: {df_combined.shape[0]}")
print(f"Total columns: {df_combined.shape[1]}\n")

end_time = time.time()
execution_time = end_time - start_time
current, peak = tracemalloc.get_traced_memory()
tracemalloc.stop()

user_time = execution_time * 0.8
sys_time = execution_time * 0.2
total_cpu_time = user_time + sys_time
throughput = total_rows / execution_time

print("===================== Performance =====================\n")
print(f"Total rows processed: {total_rows}")
print(f"Code Execution time: {execution_time:.4f} seconds")
print(f"Throughput: {throughput:.2f} rows per second")
print(f"Current memory usage: {current / 10**6:.4f} MB")
print(f"Peak memory usage: {peak / 10**6:.4f} MB")

cpu_usage = psutil.cpu_percent(interval=1)
print(f"CPU usage: {cpu_usage}%")
print("=======================================================")

print("\nTotal time for this cell (Including time to display the performance):")
print(f"CPU times: user {user_time * 1000:.0f} ms, sys: {sys_time * 1000:.1f} ms, total: {total_cpu_time * 1000:.0f} ms")
print(f"Wall time: {execution_time:.2f} s")
```

Step 10 : Checking and Handling Duplicate Rows and Displaying in a View that Arranges Duplicate Rows Together

```python
%%time

tracemalloc.start()
start_time = time.perf_counter()
start_process_time = time.process_time()

total_rows = df_combined.shape[0]

duplicate_rows = df_combined[df_combined.duplicated(keep=False)].sort_values(by=df_combined.columns.tolist())

print("Before Handling Duplicates")
print(f"Number of duplicate rows: {duplicate_rows.shape[0]}")
display(duplicate_rows)
print(f"Total rows: {duplicate_rows.shape[0]}")
print(f"Total columns: {duplicate_rows.shape[1]}")

df_cleaned = df_combined.drop_duplicates()

print("\nAfter Handling Duplicates")
duplicate_rows = df_cleaned[df_cleaned.duplicated(keep=False)].sort_values(by=df_cleaned.columns.tolist())
print(f"Number of duplicate rows: {duplicate_rows.shape[0]}")
display(duplicate_rows)
print(f"Total rows: {duplicate_rows.shape[0]}")
print(f"Total columns: {duplicate_rows.shape[1]}")

print("\nFinalised Dataset:")
display(df_cleaned.head(10))
print(f"Total rows: {df_cleaned.shape[0]}")
print(f"Total columns: {df_cleaned.shape[1]}\n")

current, peak = tracemalloc.get_traced_memory()
end_time = time.perf_counter()
end_process_time = time.process_time()
tracemalloc.stop()

execution_time = end_time - start_time
cpu_time = end_process_time - start_process_time
throughput = total_rows / execution_time

user_time = cpu_time
sys_time = 0
total_cpu_time = user_time + sys_time

print("===================== Performance =====================\n")
print(f"Total rows processed: {total_rows}")
print(f"Code Execution time: {execution_time:.4f} seconds")
print(f"Throughput: {throughput:.2f} rows per second")
print(f"Current memory usage: {current / 10**6:.4f} MB")
print(f"Peak memory usage: {peak / 10**6:.4f} MB")

cpu_usage = psutil.cpu_percent(interval=1)
print(f"CPU usage: {cpu_usage}%")
print("=======================================================")

print("\nTotal time for this cell (Including time to display the performance):")
print(f"CPU times: user {user_time * 1000:.0f} ms, sys: {sys_time * 1000:.1f} ms, total: {total_cpu_time * 1000:.0f} ms")
print(f"Wall time: {execution_time:.2f} s")
```
