```python
import matplotlib.pyplot as plt
import numpy as np

# ==========================================
# 1. Benchmark Data (From Averages)
# ==========================================
frameworks = ['Pandas', 'Dask', 'Polars']

# Data rounded for cleaner display on the charts
time_sec = [25.66, 152.32, 6.21]
cpu_usage = [59.8, 70.18, 69.36]
ram_mb = [2239.7, 4578.36, 6022.84]
throughput = [4813.0, 811.0, 22069.6]

# Custom colors for each framework
colors = ['#FF6B6B', '#FFD166', '#4ECDC4'] # Pandas (Red), Dask (Yellow), Polars (Green)

# ==========================================
# 2. Setup the 2x2 Dashboard Figure
# ==========================================
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('Performance Benchmark Comparison: Pandas vs Dask vs Polars',
             fontsize=18, fontweight='bold', y=1.02)

# ==========================================
# 3. Plotting Individual Metrics
# ==========================================

# Chart 1: Total Processing Time
axes[0, 0].bar(frameworks, time_sec, color=colors)
axes[0, 0].set_title('🕒 Total Processing Time (Lower is Better)', fontsize=14)
axes[0, 0].set_ylabel('Seconds (s)')
for i, v in enumerate(time_sec):
    axes[0, 0].text(i, v + 3, f"{v}s", ha='center', fontweight='bold', fontsize=11)

# Chart 2: CPU Usage
axes[0, 1].bar(frameworks, cpu_usage, color=colors)
axes[0, 1].set_title('🧠 Average CPU Usage (%)', fontsize=14)
axes[0, 1].set_ylabel('Percentage (%)')
axes[0, 1].set_ylim(0, 100) # Lock Y-axis to 100%
for i, v in enumerate(cpu_usage):
    axes[0, 1].text(i, v + 2, f"{v}%", ha='center', fontweight='bold', fontsize=11)

# Chart 3: Memory Usage
axes[1, 0].bar(frameworks, ram_mb, color=colors)
axes[1, 0].set_title('💾 Peak Memory Usage (Lower is Better)', fontsize=14)
axes[1, 0].set_ylabel('Megabytes (MB)')
for i, v in enumerate(ram_mb):
    axes[1, 0].text(i, v + 150, f"{v} MB", ha='center', fontweight='bold', fontsize=11)

# Chart 4: Throughput
axes[1, 1].bar(frameworks, throughput, color=colors)
axes[1, 1].set_title('⚡ Throughput (Higher is Better)', fontsize=14)
axes[1, 1].set_ylabel('Records / Second (rps)')
for i, v in enumerate(throughput):
    axes[1, 1].text(i, v + 500, f"{v} rps", ha='center', fontweight='bold', fontsize=11)

# ==========================================
# 4. Final Styling and Saving
# ==========================================
# Add gridlines behind the bars for easier reading
for ax in axes.flat:
    ax.grid(axis='y', linestyle='--', alpha=0.7)
    ax.set_axisbelow(True)

plt.tight_layout()

# Save the chart as a high-quality PNG image
plt.savefig('performance_benchmark_dashboard.png', dpi=300, bbox_inches='tight')

# Display the chart in Colab
plt.show()

print("✅ Chart successfully generated and saved as 'performance_benchmark_dashboard.png'")
```
