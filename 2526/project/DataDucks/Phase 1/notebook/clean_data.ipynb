## **Step 1: Load the Raw Data**

```python
import pandas as pd
import json
from google.colab import drive
drive.mount("/content/drive")

with open("/content/drive/MyDrive/DataDucks Project P1/raw_data.json", encoding="utf-8") as f:
    data = json.load(f)

df = pd.DataFrame(data)
print(df.shape)
print(df.dtypes)
print(df.head())
```

## **Step 2 — Inspect for issues**

```python
print(df.isnull().sum())
print(df.duplicated().sum())
```

## **Step 3 — Clean string fields**

```python
df["title"] = df["title"].str.strip().str.replace(r"\s+", " ", regex=True)
df["category"] = df["category"].str.strip().str.title()
df["excerpt"] = df["excerpt"].str.strip().str.replace(r"\s+", " ", regex=True)
df["url"] = df["url"].str.strip()

print("String cleaning done")
print(df.head())
```

## **Step 4 — Parse and standardise date**

```python
df["date"] = pd.to_datetime(df["date"], errors="coerce")
print("Unparseable dates:", df["date"].isnull().sum())
df = df.dropna(subset=["date"])
df["date"] = df["date"].dt.strftime("%Y-%m-%d")

print("Date cleaning done")
print(df["date"].head())
```

## **Step 5 — Validate URLs**

```python
before = len(df)
df = df[df["url"].str.startswith("http")]
print(f"Removed {before - len(df)} invalid URLs")
```

## **Step 6 — Final check and reset index**

```python
df = df.reset_index(drop=True)
print(f"Final record count: {len(df)}")
print(df.isnull().sum())
print(df.head())
```

## **Step 7 — Export to CSV**

```python
df.to_csv("/content/drive/MyDrive/DataDucks Project P1/cleaned_data.csv", index=False, encoding="utf-8")
print("Saved cleaned_data.csv with", len(df), "records")
```
