### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv).

### Step 3. Assign it to a variable called chipo.

```python
file_path = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv'
chipo = pd.read_csv(file_path, delimiter='\t')
```

Display the dataframe from dataset.

```python
chipo #to displays dataframe
chipo.head(15) #display dataframe for 15 item
```

### Step 4. How many products cost more than $10.00?

To perform comparison of the product cost, we have to convert item_price to float data type and remove the '$'

```python
chipo['item_price'] = chipo['item_price'].str.replace('$','').astype(float)
```

Compare item_price and assign the value to a new variable called pricey_item

```python
pricey_item = chipo[chipo['item_price'] > 10.00]
count = len(pricey_item)
```

Display the pricey item and the quantity of product that cost more than $10.00

```python
print(" The number of product that cost more than $10.00 is",count)
```

```python
pricey_item #display the pricey item
```

### Step 5. What is the price of each item?
###### print a data frame with only two columns item_name and item_price

```python
price = chipo.groupby('item_name')['item_price'].first().reset_index()
print(price)
```

### Step 6. Sort by the name of the item

```python
sorted_name = chipo.sort_values(by = 'item_name').reset_index()
sorted_name
```

### Step 7. What was the quantity of the most expensive item ordered?

Find for the highest item price ordered and its quantity

```python
highest_item_price = chipo['item_price'].max()
quantity_highest = chipo[chipo['item_price'] == highest_item_price]['quantity'].sum()
print(f"The quantity of the most expensive item ordered is {quantity_highest} items")
print()
most_expensive = chipo[chipo['item_price'] == highest_item_price]
most_expensive #display dataframe for the highest item
```

### Step 8. How many times was a Veggie Salad Bowl ordered?

```python
veggieSalad = chipo[chipo['item_name'] == 'Veggie Salad Bowl']
veggieOrder = len(veggieSalad)
print(f"Total order of Veggie Salad Bowl is {veggieOrder} times")
print()
veggieSalad #display the dataframe of Veggie Salad Bold
```

### Step 9. How many times did someone order more than one Canned Soda?

```python
cannedSoda = chipo[(chipo['item_name'] == 'Canned Soda') & (chipo['quantity'] > 1)]
sodaOrder = len(cannedSoda)
print(f"The total order of someone who orders more than one canned of soda is {sodaOrder} times")
print()
cannedSoda #display the dataframe
```
