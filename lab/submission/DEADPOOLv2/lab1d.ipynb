# Lab 1d - Filtering and Sorting Data

Please use the [**Chipotle Sales dataset**](https://raw.githubusercontent.com/drshahizan/dataset/main/pandas/chipotle.tsv) for this exercise.

### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv).

### Step 3. Assign it to a variable called chipo.

```python
chipo = pd.read_table('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv')

chipo
```

### Step 4. How many products cost more than $10.00?

checking the datatype for column "item_price", we notice that the dtype of the column is an object.

```python
chipo['item_price']
```

changing the datatype for column "item_price" from object to float64

```python
chipo['item_price'] = chipo['item_price'].str.replace('$', '')
chipo['item_price'] = chipo['item_price'].astype(float)
chipo['item_price']
```

determining the quantity of products that cost more than $10.00


```python
chipo_ten = np.count_nonzero(chipo['item_price'] > 10)
print("Quantity of products ordered that are more than $10.00: ", chipo_ten)
```

### Step 5. What is the price of each item?
###### print a data frame with only two columns item_name and item_price

printed below is a df of only 2 columns, "item_name" and also "item_price", showcasing each item with respective price

```python
chipotle = chipo.groupby('item_name')
chipotle = chipotle['item_price'].sum() / chipotle['quantity'].sum()
chipotle = pd.DataFrame({'item_name': chipotle.index, 'price_per_item': chipotle.values})
chipotle['price_per_item'] = chipotle['price_per_item'].round(2)
chipotle
```

### Step 6. Sort by the name of the item

below is the sorted df by "item_name" in ascending order, with new assigned index

```python
chipo.sort_values(by='item_name', ascending=True).reset_index(drop=True)
```

### Step 7. What was the quantity of the most expensive item ordered?

below is code to find the most expensive order, assigned the values to a new df called "chipo_exp"

```python
chipo_exp = chipo[chipo['item_price']==chipo['item_price'].max()]
chipo_exp
for i in chipo_exp['quantity']:
    if i!=0:
        print("The quantity of the most expensive order is: ",i)
```

to find the quantity of the most expensive item, ordered, we first need a df of price by item

```python
chipotle
```

then we filter out the most expensive ones

```python
chipo_mahal = chipotle[chipotle.price_per_item == chipotle.price_per_item.max()]
chipo_mahal
```

based on the previously obtained information, we can work on the whole order list to count the quantity of the most expensive item

```python
chipo_mahal = chipo[chipo['item_name'].isin(['Steak Salad Bowl'])]
chipo_mahal
```

```python
steak_price_mahal = chipo_mahal
steak_price_mahal['per_steak'] = steak_price_mahal['item_price'] / steak_price_mahal['quantity']
max_per_steak = steak_price_mahal['per_steak'].max()
steak_price_mahal = steak_price_mahal[steak_price_mahal['per_steak'] == max_per_steak]
steak_price_mahal
```

summing the quantity of items in each order

```python
steak_price_mahal['quantity'].sum()
```

```python
print("The quantity of the most expensive items ordered is: " , steak_price_mahal['quantity'].sum())
```
