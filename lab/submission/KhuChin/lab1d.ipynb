# Lab 1d - Filtering and Sorting Data

Please use the [**Chipotle Sales dataset**](https://raw.githubusercontent.com/drshahizan/dataset/main/pandas/chipotle.tsv) for this exercise.

### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv).

```python
file_path = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv"
```

### Step 3. Assign it to a variable called chipo.

```python
chipo = pd.read_csv(file_path, sep='\t')
#displaying dataset
chipo
```

### Step 4. How many products cost more than $10.00?

```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

# First, convert the 'item_price' column to a string to remove the dollar sign
chipo['item_price'] = chipo['item_price'].astype(str)

# Remove the dollar sign and convert to float
chipo['item_price'] = chipo['item_price'].str.replace('$', '').astype(float)

# Use boolean indexing to filter for products that cost more than $10.00
luxury_products = chipo[chipo['item_price'] > 10.00]

# Count the number of expensive products using the 'count' method
luxury_products = luxury_products['item_name'].count()

# Print the result
print("Number of products that cost more than $10.00 is :", luxury_products,"items")
```

### Step 5. What is the price of each item?
###### print a data frame with only two columns item_name and item_price

```python
# Select only the item_name and item_price columns
chipo_selected = chipo[["item_name", "item_price"]]

# Print the selected dataframe
chipo_selected
```

### Step 6. Sort by the name of the item

```python
sort_item_name=chipo
sort_item_name.sort_values(by='item_name')
```

### Step 7. What was the quantity of the most expensive item ordered?

```python
# Find the most expensive item ordered by sorting the dataframe by the item_price column in descending order and selecting the first row
chipo_sorted = chipo.sort_values(by="item_price", ascending=False)
most_expensive_item = chipo_sorted.iloc[0]
most_expensive_item
```

### Step 8. How many times was a Veggie Salad Bowl ordered?

```python
# Filter the dataframe to keep only the rows where the item_name is Veggie Salad Bowl
chipo_veggie = chipo[chipo["item_name"] == "Veggie Salad Bowl"]

# Count the number of times Veggie Salad Bowl was ordered by summing up the quantity column
count = chipo_veggie["quantity"].sum()

# Print the answer
print(f"Veggie Salad Bowl was ordered {count} times.")
```

### Step 9. How many times did someone order more than one Canned Soda?

```python
# Filter the dataframe to keep only the rows where the item_name is Canned Soda and the quantity is more than one
chipo_soda = chipo[(chipo["item_name"] == "Canned Soda") & (chipo["quantity"] > 1)]

# Count the number of times someone ordered more than one Canned Soda by counting the number of rows in the filtered dataframe
count = len(chipo_soda)

# Print the answer
print(f"Someone ordered more than one Canned Soda {count} times.")
```
